```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
file_path = '1728589899591.csv'
df = pd.read_csv(file_path)

# Sorting the data by 'Asset ID' and 'Disposition'
sorted_df = df.sort_values(by=['Asset ID', 'Disposition'])

# Counting the number of items per 'Disposition'
disposition_counts = sorted_df['Disposition'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(disposition_counts.index, disposition_counts.values, color='skyblue')

# Adding numbers on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

plt.title('Number of Items per Disposition')
plt.xlabel('Disposition')
plt.ylabel('Number of Items')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'ICLF Weekly Stats (Responses) - Form Responses 1.csv'
data = pd.read_csv(file_path)

# Convert the 'Week of Data Collection (Friday)' column to datetime format
data['Week of Data Collection (Friday)'] = pd.to_datetime(data['Week of Data Collection (Friday)'])

# Extract 'Year' and 'Week' columns
data['Year'] = data['Week of Data Collection (Friday)'].dt.year
data['Week'] = data['Week of Data Collection (Friday)'].dt.strftime('%W').astype(int)

# Filter data to include only weeks 35 and later
data = data[data['Week'] >= 35]

# Group by 'Year' and 'Week', summing the 'Returned to Owner' counts
weekly_stats = data.groupby(['Year', 'Week'])['Returned to Owner'].sum().reset_index()

# Separate data for last year and this year
current_year = pd.Timestamp.now().year
last_year = weekly_stats[weekly_stats['Year'] == (current_year - 1)]
this_year = weekly_stats[weekly_stats['Year'] == current_year]

# Merge last year's and this year's data on 'Week'
comparison_stats = pd.merge(last_year, this_year, on='Week', how='outer', suffixes=('_last_year', '_this_year')).fillna(0)

# Sort by 'Week'
comparison_stats = comparison_stats.sort_values(by='Week')

# Plot the comparison
plt.figure(figsize=(12, 6))
plt.bar(comparison_stats['Week'] - 0.2, comparison_stats['Returned to Owner_last_year'], width=0.4, color='skyblue', label=f'Last Year')
plt.bar(comparison_stats['Week'] + 0.2, comparison_stats['Returned to Owner_this_year'], width=0.4, color='salmon', label=f'This Year')

# Add labels, title, and customize ticks
plt.xlabel('Week')
plt.ylabel('Number of Items Returned to Owner')
plt.title('Comparison of Returned to Owner Items by Week (Starting from Week 35, Last Year vs. This Year)')
plt.xticks(comparison_stats['Week'])
plt.legend()

# Annotate the bars with the values
for i, (last, this, week) in enumerate(zip(comparison_stats['Returned to Owner_last_year'], comparison_stats['Returned to Owner_this_year'], comparison_stats['Week'])):
    plt.annotate(f'{int(last)}', xy=(week - 0.2, last), ha='center', va='bottom', color='black')
    plt.annotate(f'{int(this)}', xy=(week + 0.2, this), ha='center', va='bottom', color='black')

# Show the plot
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'ICLF Weekly Stats (Responses) - Form Responses 1.csv'
data = pd.read_csv(file_path)

# Convert the 'Week of Data Collection (Friday)' column to datetime format
data['Week of Data Collection (Friday)'] = pd.to_datetime(data['Week of Data Collection (Friday)'])

# Extract 'Year' and 'Week' columns
data['Year'] = data['Week of Data Collection (Friday)'].dt.year
data['Week'] = data['Week of Data Collection (Friday)'].dt.strftime('%W').astype(int)

# Filter data to include only weeks 35 and later
data = data[data['Week'] >= 35]

# Group by 'Year' and 'Week', summing the required fields
weekly_stats = data.groupby(['Year', 'Week'])[['Returned to Owner', 'Items Available', 'Donated']].sum().reset_index()

# Separate data for last year and this year
current_year = pd.Timestamp.now().year
last_year = weekly_stats[weekly_stats['Year'] == (current_year - 1)]
this_year = weekly_stats[weekly_stats['Year'] == current_year]

# Merge last year's and this year's data on 'Week'
comparison_stats = pd.merge(last_year, this_year, on='Week', how='outer', suffixes=('_last_year', '_this_year')).fillna(0)

# Sort by 'Week'
comparison_stats = comparison_stats.sort_values(by='Week')

# Define a function to plot comparisons for a specific column
def plot_comparison(column, color1, color2):
    plt.figure(figsize=(12, 6))
    plt.bar(comparison_stats['Week'] - 0.2, comparison_stats[f'{column}_last_year'], width=0.4, color=color1, label=f'Last Year')
    plt.bar(comparison_stats['Week'] + 0.2, comparison_stats[f'{column}_this_year'], width=0.4, color=color2, label=f'This Year')

    # Add labels, title, and customize ticks
    plt.xlabel('Week')
    plt.ylabel(f'Number of {column.replace("_", " ").title()}')
    plt.title(f'Comparison of {column.replace("_", " ").title()} by Week (Starting from Week 35, Last Year vs. This Year)')
    plt.xticks(comparison_stats['Week'])
    plt.legend()

    # Annotate the bars with the values
    for i, (last, this, week) in enumerate(zip(comparison_stats[f'{column}_last_year'], comparison_stats[f'{column}_this_year'], comparison_stats['Week'])):
        plt.annotate(f'{int(last)}', xy=(week - 0.2, last), ha='center', va='bottom', color='black')
        plt.annotate(f'{int(this)}', xy=(week + 0.2, this), ha='center', va='bottom', color='black')

    # Show the plot
    plt.tight_layout()
    plt.show()

# Plot each comparison
plot_comparison('Returned to Owner', 'skyblue', 'salmon')
plot_comparison('Items Available', 'lightgreen', 'darkgreen')
plot_comparison('Donated', 'lightcoral', 'darkred')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'ICLF Weekly Stats (Responses) - Form Responses 1.csv'
data = pd.read_csv(file_path)

# Convert the 'Week of Data Collection (Friday)' column to datetime format
data['Week of Data Collection (Friday)'] = pd.to_datetime(data['Week of Data Collection (Friday)'])

# Extract 'Year' and 'Week' columns
data['Year'] = data['Week of Data Collection (Friday)'].dt.year
data['Week'] = data['Week of Data Collection (Friday)'].dt.strftime('%W').astype(int)

# Filter data to include only weeks 35 and later
data = data[data['Week'] >= 35]

# Group by 'Year' and 'Week', summing the necessary fields
weekly_stats = data.groupby(['Year', 'Week'])[['Returned to Owner', 'Items Available']].sum().reset_index()

# Calculate the "Returned to Owner" percentage
weekly_stats['Returned_to_Owner_Percentage'] = (weekly_stats['Returned to Owner'] / weekly_stats['Items Available']) * 100

# Separate data for last year and this year
current_year = pd.Timestamp.now().year
last_year_stats = weekly_stats[weekly_stats['Year'] == (current_year - 1)]
this_year_stats = weekly_stats[weekly_stats['Year'] == current_year]

# Merge last year's and this year's data on 'Week' for comparison
comparison_stats = pd.merge(last_year_stats[['Week', 'Returned_to_Owner_Percentage']],
                            this_year_stats[['Week', 'Returned_to_Owner_Percentage']],
                            on='Week', how='outer', suffixes=('_last_year', '_this_year')).fillna(0)

# Sort by 'Week'
comparison_stats = comparison_stats.sort_values(by='Week')

# Plot the comparison
plt.figure(figsize=(12, 6))
plt.plot(comparison_stats['Week'], comparison_stats['Returned_to_Owner_Percentage_last_year'], marker='o', color='skyblue', label='Last Year')
plt.plot(comparison_stats['Week'], comparison_stats['Returned_to_Owner_Percentage_this_year'], marker='o', color='salmon', label='This Year')

# Add labels, title, and customize ticks
plt.xlabel('Week')
plt.ylabel('Returned to Owner Percentage (%)')
plt.title('Comparison of Returned to Owner Percentage by Week (Starting from Week 35, Last Year vs. This Year)')
plt.xticks(comparison_stats['Week'])
plt.legend()

# Annotate the lines with the values
for i, (last, this, week) in enumerate(zip(comparison_stats['Returned_to_Owner_Percentage_last_year'], 
                                           comparison_stats['Returned_to_Owner_Percentage_this_year'], 
                                           comparison_stats['Week'])):
    plt.annotate(f'{last:.1f}%', xy=(week, last), ha='center', va='bottom', color='skyblue')
    plt.annotate(f'{this:.1f}%', xy=(week, this), ha='center', va='bottom', color='salmon')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'ICLF Weekly Stats (Responses) - Form Responses 1.csv'
data = pd.read_csv(file_path)

# Convert the 'Week of Data Collection (Friday)' column to datetime format
data['Week of Data Collection (Friday)'] = pd.to_datetime(data['Week of Data Collection (Friday)'])

# Extract 'Year' and 'Month' from the 'Week of Data Collection (Friday)' column
data['Year'] = data['Week of Data Collection (Friday)'].dt.year
data['Month'] = data['Week of Data Collection (Friday)'].dt.month

# Group by 'Year' and 'Month', summing the necessary fields
monthly_stats = data.groupby(['Year', 'Month'])[['Returned to Owner', 'Items Available']].sum().reset_index()

# Calculate the "Returned to Owner" percentage for each month
monthly_stats['Returned_to_Owner_Percentage'] = (monthly_stats['Returned to Owner'] / monthly_stats['Items Available']) * 100

# Separate data for last year and this year
current_year = pd.Timestamp.now().year
last_year_stats = monthly_stats[monthly_stats['Year'] == (current_year - 1)]
this_year_stats = monthly_stats[monthly_stats['Year'] == current_year]

# Merge last year's and this year's data on 'Month' for comparison
comparison_stats = pd.merge(last_year_stats[['Month', 'Returned_to_Owner_Percentage']],
                            this_year_stats[['Month', 'Returned_to_Owner_Percentage']],
                            on='Month', how='outer', suffixes=('_last_year', '_this_year')).fillna(0)

# Sort by 'Month'
comparison_stats = comparison_stats.sort_values(by='Month')

# Plot the comparison
plt.figure(figsize=(12, 6))
plt.plot(comparison_stats['Month'], comparison_stats['Returned_to_Owner_Percentage_last_year'], marker='o', color='skyblue', label='Last Year')
plt.plot(comparison_stats['Month'], comparison_stats['Returned_to_Owner_Percentage_this_year'], marker='o', color='salmon', label='This Year')

# Add labels, title, and customize ticks
plt.xlabel('Month')
plt.ylabel('Returned to Owner Percentage (%)')
plt.title('Comparison of Returned to Owner Percentage by Month (Last Year vs. This Year)')
plt.xticks(range(1, 13))  # Months 1 through 12
plt.legend()

# Annotate the lines with the values
for last, this, month in zip(comparison_stats['Returned_to_Owner_Percentage_last_year'], 
                             comparison_stats['Returned_to_Owner_Percentage_this_year'], 
                             comparison_stats['Month']):
    plt.annotate(f'{last:.1f}%', xy=(month, last), ha='center', va='bottom', color='skyblue')
    plt.annotate(f'{this:.1f}%', xy=(month, this), ha='center', va='bottom', color='salmon')

# Show the plot
plt.tight_layout()
plt.show()
```
